```python
import networkx as nx
import matplotlib.pyplot as plt
import matlab as plta
import numpy as np
```

Se ingresa por teclado la cantidad de numeros que representen la longitud de la contraseña.

```python
longitudContraseña = input()
```

```python
def contraseña(numero):
    numero = 10**numero
    return numero
```

Se puede ver la cantidad de posibilidades dependiendo de la longitud de numeros, si por ejemplo, son 4 digitos, el numero de posibilidades es 10^4 = 10000

```python
contraseña(int(longitudContraseña))
```

Pero ahora vamos a verlo en un termino en cuanto a grafos, si tenemos 4 nodos, y cada uno esta conectado por 3 aristas, entonces podemos contruir una matriz de adyacencia de 10 x 10 (la cantidad de digitos que puede recibir cada valor del 0 al 9) y ver la cantidad de cadenas que hay entre 2 nodos.

```python
matriz = np.ones((10,10))
```

```python
matrizFinal = np.linalg.matrix_power(matriz, int(longitudContraseña)-1)
```

```python
matrizFinal
```

ahora sumamos cada valor de la matriz y ahi nos daria la cantidad de posibilidades que hay para nuestra contraseña

```python
contraseña = np.sum(matrizFinal)
```

```python
contraseña
```

y aca podemos ver nuestro grafo donde tenemos los 10 valores y cada uno conectado entre si.

```python
grafo = nx.Graph()


for d in range(10):
    grafo.add_node(d)
    
    
for d1 in range(10):
    for d2 in range(10):
        grafo.add_edge(d1, d2)
        
        
nx.draw(grafo, with_labels=True)
plt.show()
```
