```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo vacío
G = nx.Graph()

# Crear una lista de todas las combinaciones posibles
combinations = [str(i).zfill(4) for i in range(10000)]

# Agregar cada combinación como un nodo en el grafo
for combination in combinations:
    G.add_node(combination)

# Agregar bordes entre las combinaciones que solo difieren en un dígito
for i in range(len(combinations)):
    for j in range(i+1, len(combinations)):
        combination1 = combinations[i]
        combination2 = combinations[j]
        diff_count = sum(c1 != c2 for c1, c2 in zip(combination1, combination2))
        if diff_count == 1:
            G.add_edge(combination1, combination2)

# Definir el diseño del grafo
pos = nx.spring_layout(G)

# Dibujar el grafo
nx.draw_networkx(G, pos, with_labels=False, node_size=5)

# Mostrar el grafo
plt.axis('off')
plt.show()
```
